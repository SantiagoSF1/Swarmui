````python
# Descargar Videos de Drive

#@markdown ### **Descargar video de Drive**
#@markdown Este script permite descargar videos desde una carpeta compartida de Google Drive. El usuario puede seleccionar un archivo y descargarlo en una carpeta local o en Google Drive.
#@markdown Aseg√∫rate de tener los permisos necesarios para acceder a la carpeta y descargar los archivos.

# M√≥dulos necesarios
from google.colab import auth
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseDownload
import io, os, re, requests
from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets

# Autenticaci√≥n para Google Drive
auth.authenticate_user()
drive_service = build('drive', 'v3')

guardar_en_drive = False  # @param {type:"boolean"}
#@markdown El proceso incluye una interfaz interactiva para seleccionar el archivo, una barra de progreso personalizada y opciones para confirmar o cancelar la descarga.
folder_link = "https://drive.google.com/drive/folders/1z6Ky2wKkA5gYAQkKXVKZMhonjjbhOGHh"  # @param {type:"string"}

# Rutas de las carpetas
local_folder = "/content/FFmpeg/videos"
drive_folder = "/content/drive/MyDrive/videos"
os.makedirs(local_folder, exist_ok=True)
if guardar_en_drive: os.makedirs(drive_folder, exist_ok=True)

# Funci√≥n para extraer el ID de la carpeta
def extract_folder_id_from_url(url):
    folder_id = None
    patterns = [r"/folders/([a-zA-Z0-9_-]+)"]
    for pattern in patterns:
        match = re.search(pattern, url)
        if match: folder_id = match.group(1); break
    return folder_id

# Funci√≥n para listar archivos en la carpeta
def list_files_in_folder(folder_id):
    files = []
    page_token = None
    while True:
        response = drive_service.files().list(q=f"'{folder_id}' in parents", spaces='drive', fields='nextPageToken, files(id, name)', pageToken=page_token).execute()
        files.extend(response.get('files', []))
        page_token = response.get('nextPageToken', None)
        if page_token is None: break
    return files

# Extraer el ID de la carpeta
folder_id = extract_folder_id_from_url(folder_link)
if not folder_id: raise Exception("‚ùå No se pudo extraer el ID de la carpeta del enlace proporcionado.")

# Listar archivos en la carpeta
files = list_files_in_folder(folder_id)
if not files: raise Exception("‚ùå No se encontraron archivos en la carpeta.")

# Crear men√∫ desplegable para seleccionar archivo
file_names = [file['name'] for file in files]
file_selector = widgets.Dropdown(options=file_names, description='')
display(file_selector)
display_outside_box("Selecciona el video a descargar", Wide="265px")

# Funci√≥n para descargar archivo desde Google Drive
def download_from_drive(file_id, nombre):
    file_metadata = drive_service.files().get(fileId=file_id, fields="size, name").execute()
    file_size = int(file_metadata['size'])
    file_name = file_metadata.get('name', nombre)

    # Crear la barra de progreso personalizada
    display_outside_box("Descargando el video seleccionado", Wide="300px")
    progress = display_ProgressBarAnimated(
        total=file_size,
        text="",
        width=320,
        height=30,
        color1="#3498db",  # azul
        color2="#9b59b6")  # morado

    # Configurar la solicitud para descargar el archivo
    request = drive_service.files().get_media(fileId=file_id)
    output_path = os.path.join(local_folder, file_name)
    fh = io.FileIO(output_path, 'wb')
    downloader = MediaIoBaseDownload(fh, request)
    done = False
    while done is False:
        status, done = downloader.next_chunk()
        progress.update(int(status.progress() * file_size))
    if guardar_en_drive:
        drive_path = os.path.join(drive_folder, file_name)
        with open(drive_path, 'wb') as f:
            with open(output_path, 'rb') as temp_file:
                f.write(temp_file.read())

    # Finalizar la barra de progreso
    progress.complete("Descarga completada")
    clear_output()
    display_outside_box(f"Video descargado", Wide="250px")

# Funci√≥n para manejar la confirmaci√≥n
def on_confirm_button_click(b):
    clear_output()
    try:
        selected_file_name = file_selector.value
        selected_file_id = next(file['id'] for file in files if file['name'] == selected_file_name)
        download_from_drive(selected_file_id, selected_file_name)
    except Exception as e:
        display_outside_box(f"Error: {str(e)}", "#F9AB00", "#ad7900")

# Funci√≥n para manejar la cancelaci√≥n
def on_cancel_button_click(b):
    clear_output()
    display_outside_box("Descarga cancelada", "#DD0700", Wide="250px")
# Crear botones para confirmar o cancelar
confirm_button = widgets.Button(description="Confirmar Descarga", button_style='primary', layout=widgets.Layout(width='150px', height='40px'))
cancel_button = widgets.Button(description="Cancelar Descarga", button_style='danger', layout=widgets.Layout(width='150px', height='40px'))
confirm_button.style.button_color, cancel_button.style.button_color = "#292929", "#292929"
confirm_button.on_click(on_confirm_button_click)
cancel_button.on_click(on_cancel_button_click)
display(widgets.HBox([confirm_button, cancel_button]))

# Fin
````

````python
# Extracci√≥n de Frames

import os
import subprocess
import shutil
import ipywidgets as widgets
from IPython.display import display, clear_output

#@markdown ### **Extracci√≥n de Frames**
#@markdown Este script permite extraer frames de videos mp4 utilizando FFmpeg con aceleraci√≥n por GPU. El usuario puede seleccionar un video de una lista y elegir cu√°ntos frames por segundo `FPS` desea extraer.
#@markdown Los frames extra√≠dos se guardar√°n en una carpeta `frames` dentro de tu proyecto, con opci√≥n de dividirlos en subcarpetas.
fps_video = 1  # @param {type: "slider", min: 0.2, max: 2, step: 0.2}

#@markdown El proceso incluye una interfaz interactiva para seleccionar el video, una barra de progreso personalizada y opciones para confirmar o cancelar la extracci√≥n.

ruta_drive = "drive/MyDrive/"
video_folder = "/content/FFmpeg/videos"
Frames_e = os.path.join(ruta_drive, "Capture")
os.makedirs(Frames_e, exist_ok=True)

videos_disponibles = [f for f in os.listdir(video_folder) if f.endswith('.mp4')]
cancel_extraction = False

# Opci√≥n para seleccionar el video a procesar
if not videos_disponibles:
    display_outside_box("No se encontraron videos", "#DD0700")
else:
    seleccion_video = widgets.Dropdown(options=videos_disponibles, description='Selecciona:', disabled=False)
    display(seleccion_video)

    # Opci√≥n para dividir los frames
    dividir_frames = widgets.Dropdown(options=["No", 1, 2, 4, 8], description='Dividir:', disabled=False)
    display(dividir_frames)
    display_outside_box("Selecciona el video a procesar", Wide="265px")

    # Funci√≥n para extraer frames con FFmpeg y GPU
    def extract_frames(video_path, output_folder, fps, dividir):
        global cancel_extraction
        video_name = os.path.splitext(os.path.basename(video_path))[0]  # sin reemplazar espacios

        # Carpeta final de frames
        subcarpeta_frames = os.path.join(output_folder, video_name)
        os.makedirs(subcarpeta_frames, exist_ok=True)

        # Obtener la duraci√≥n del video para calcular el progreso
        ffprobe_command = [
            'ffprobe', '-v', 'error', '-show_entries', 'format=duration',
            '-of', 'default=noprint_wrappers=1:nokey=1', video_path
        ]
        duration = float(subprocess.run(ffprobe_command, stdout=subprocess.PIPE, text=True).stdout.strip())

        # Crear la barra de progreso
        display_outside_box("Extrayendo frames del video...", Wide="300px")
        progress_bar = display_ProgressBarAnimated(
            total=int(duration),
            text="",
            width=320,
            height=30,
            color1="#3498db",
            color2="#9b59b6")

        # Comando FFmpeg optimizado GPU + CPU
        ffmpeg_command = [
            'ffmpeg',
            '-hwaccel', 'cuda',        # Usar aceleraci√≥n GPU
            '-threads', '0',           # Usar todos los hilos CPU
            '-i', video_path,
            '-vf', f'fps={fps}',
            '-q:v', '1',               # Calidad de los frames
            '-preset', 'fast',         # Optimiza velocidad
            '-f', 'image2',
            os.path.join(subcarpeta_frames, f"frame %04d.jpg")
        ]

        # Ejecutar FFmpeg y actualizar la barra de progreso
        try:
            process = subprocess.Popen(ffmpeg_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, universal_newlines=True)
            while True:
                output = process.stderr.readline()
                if output == '' and process.poll() is not None:
                    break
                if output:
                    # Extraer el tiempo actual del frame procesado
                    if "frame=" in output and "time=" in output:
                        time_str = output.split("time=")[1].split(" ")[0]
                        h, m, s = map(float, time_str.split(':'))
                        current_time = h * 3600 + m * 60 + s
                        progress_bar.update(int(current_time) - progress_bar.current)
            process.wait()
            if process.returncode == 0:
                clear_output(wait=True)  # Borra la salida anterior
                progress_bar.complete("Extracci√≥n finalizada")
                if dividir != "No":
                    dividir_frames_en_subcarpetas(subcarpeta_frames, dividir)

                display_outside_box("Carpeta guardada", Wide="265px")

            else:
                display_outside_box("Error durante la extracci√≥n", "#DD0700")
        except KeyboardInterrupt:
            process.terminate()
            display_outside_box("Extracci√≥n cancelada", "#DD0700")

    # Funci√≥n para dividir los frames en subcarpetas con barra de progreso
    def dividir_frames_en_subcarpetas(carpeta_frames, dividir):
        frames = sorted([f for f in os.listdir(carpeta_frames) if f.endswith('.jpg')])
        total_frames = len(frames)
        frames_por_subcarpeta = total_frames // dividir

        # Crear la barra de progreso para la divisi√≥n de frames
        display_outside_box("Dividiendo frames del video...", Wide="300px")
        progress_bar_div = display_ProgressBarAnimated(
            total=total_frames,
            text="",
            width=320,
            height=30,
            color1="#3498db",
            color2="#9b59b6")

        for i in range(dividir):
            subcarpeta = os.path.join(carpeta_frames, f"frames {i+1}")
            os.makedirs(subcarpeta, exist_ok=True)
            inicio = i * frames_por_subcarpeta
            fin = inicio + frames_por_subcarpeta if i < dividir - 1 else total_frames
            for frame in frames[inicio:fin]:
                os.rename(os.path.join(carpeta_frames, frame), os.path.join(subcarpeta, frame))
                progress_bar_div.update(1)  # Actualizar la barra de progreso

        clear_output()
        progress_bar_div.complete("Divisi√≥n finalizada")
        display_outside_box("Extracci√≥n finalizada", Wide="265px")
        display_outside_box("Divisi√≥n finalizada", Wide="265px")

    # Funci√≥n para manejar la confirmaci√≥n
    def on_start_button_click(b):
        global cancel_extraction
        cancel_extraction = False
        if not seleccion_video.value:
            clear_output()
            display_outside_box("Selecciona un video", "#DD0700")
            return
        video_path = os.path.join(video_folder, seleccion_video.value)
        if not os.path.exists(video_path):
            clear_output()
            display_outside_box("El video no existe", "#DD0700")
            return
        clear_output()
        extract_frames(video_path, Frames_e, fps_video, dividir_frames.value)

    # Funci√≥n para manejar la cancelaci√≥n
    def on_cancel_button_click(b):
        global cancel_extraction
        cancel_extraction = True
        clear_output()
        display_outside_box("Extracci√≥n cancelada", "#DD0700")

    # Mostrar los botones para confirmar o cancelar
    start_button = widgets.Button(description="Confirmar Extracci√≥n", button_style='primary', layout=widgets.Layout(width='150px', height='45px'), style={'button_color': '#292929'})
    cancel_button = widgets.Button(description="Cancelar Extracci√≥n", button_style='primary', layout=widgets.Layout(width='150px', height='45px'), style={'button_color': '#292929'})
    start_button.on_click(on_start_button_click)
    cancel_button.on_click(on_cancel_button_click)
    display(widgets.HBox([start_button, cancel_button]))

# Fin
````

````python

# Clasificador Autom√°tico Large Tagger

#@markdown ### **Buscar personaje**
#@markdown Este script utiliza el modelo **WaifuDiffusion** para analizar im√°genes y generar etiquetas autom√°ticamente.
#@markdown Incluye soporte para:
#@markdown Detecci√≥n de **personajes** y etiquetas **generales** como ropa, expresiones y objetos. Copia autom√°tica de im√°genes similares.
utilizar_gpu = False  #@param {type:"boolean"}

#@markdown Interfaz gr√°fica con **Gradio** para cargar im√°genes, procesarlas, visualizar etiquetas y editarlas f√°cilmente.
#@markdown Los resultados se muestran en formato legible.

!pip install --upgrade gradio pandas -q
if utilizar_gpu:
    !pip install --upgrade onnxruntime-gpu -q
else:
    !pip install --upgrade onnxruntime -q

import os
import shutil
import numpy as np
import pandas as pd
from PIL import Image
import gradio as gr
import onnxruntime as rt

# Clonar repositorio de tags en la carpeta indicada
if not os.path.exists("/content/WaifuDiffusion/extension/tagcomplete"):
    !git clone https://github.com/DominikDoom/a1111-sd-webui-tagcomplete.git /content/WaifuDiffusion/extension/tagcomplete

# Crear carpeta model
model_dir = "/content/WaifuDiffusion/model_mini"
os.makedirs(model_dir, exist_ok=True)

# Variables para las rutas visualizadas
model_path = "/content/WaifuDiffusion/model_mini/model.onnx"
labels_path = "/content/WaifuDiffusion/model_mini/selected_tags.csv"

# Descargar modelo y etiquetas si no existen
if not os.path.exists(model_path):
    os.system(f"wget -q https://huggingface.co/SmilingWolf/wd-vit-tagger-v3/resolve/main/model.onnx -O {model_path}")
if not os.path.exists(labels_path):
    os.system(f"wget -q https://huggingface.co/SmilingWolf/wd-vit-tagger-v3/resolve/main/selected_tags.csv -O {labels_path}")

kaomojis = ["0_0","(o)_(o)","+_+","+-_","._.","<o>_<o>","<|>_<|>","=_=",
            ">_<","3_3","6_9",">_o","@_@","^_^","o_o","u_u","x_x","|_|","||_||"]

def load_labels():
    df = pd.read_csv(labels_path)
    names = df["name"].map(lambda x: x.replace("_"," ") if x not in kaomojis else x).tolist()
    rating_idx = list(np.where(df["category"]==9)[0])
    general_idx = list(np.where(df["category"]==0)[0])
    character_idx = list(np.where(df["category"]==4)[0])
    return names, rating_idx, general_idx, character_idx

tag_names, rating_idx, general_idx, character_idx = load_labels()

class Tagger:
    def __init__(self, model_path, utilizar_gpu=True):
        if utilizar_gpu:
            providers = ["CUDAExecutionProvider", "CPUExecutionProvider"]
        else:
            providers = ["CPUExecutionProvider"]
        self.model = rt.InferenceSession(model_path, providers=providers)
        _, h, w, _ = self.model.get_inputs()[0].shape
        self.size = h

    def prepare(self, image):
        canvas = Image.new("RGBA", image.size, (255,255,255))
        canvas.alpha_composite(image)
        image = canvas.convert("RGB")
        max_dim = max(image.size)
        pad_left = (max_dim-image.size[0])//2
        pad_top = (max_dim-image.size[1])//2
        padded = Image.new("RGB",(max_dim,max_dim),(255,255,255))
        padded.paste(image,(pad_left,pad_top))
        if max_dim!=self.size:
            padded = padded.resize((self.size,self.size), Image.BICUBIC)
        arr = np.asarray(padded,dtype=np.float32)[:,:,::-1]
        return np.expand_dims(arr,0)

    def predict(self, image, general_thresh=0.35, character_thresh=0.85):
        img = self.prepare(image)
        input_name = self.model.get_inputs()[0].name
        preds = self.model.run(None,{input_name:img})[0][0].astype(float)

        characters = {tag_names[i]: preds[i] for i in character_idx if preds[i]>character_thresh}
        sorted_char = sorted(characters.items(), key=lambda x:x[1], reverse=True)

        general = {tag_names[i]: preds[i] for i in general_idx if preds[i]>general_thresh}
        sorted_general = sorted(general.items(), key=lambda x:x[1], reverse=True)

        combined = sorted_char + sorted_general
        sorted_str = ", ".join([x[0] for x in combined])
        return sorted_str, combined

predictor = Tagger(model_path, utilizar_gpu=utilizar_gpu)

# Ruta de referencia
reference_path = "/content/WaifuDiffusion/Imagen/Reference.txt"
os.makedirs("/content/WaifuDiffusion/Imagen", exist_ok=True)

# Copia im√°genes en ruta definida con project_file
similar_dir = os.path.join(project_file, "Imagen")
os.makedirs(similar_dir, exist_ok=True)

# Procesar todas las im√°genes de una subcarpeta y copiar similares
def buscar_personaje(subfolder, general_thresh, character_thresh, min_similares):
    if not os.path.exists(subfolder):
        return f"‚ö†Ô∏è Subcarpeta no existe: {subfolder}"
    if not os.path.exists(reference_path):
        return f"‚ö†Ô∏è Reference.txt no encontrado"

    with open(reference_path, "r", encoding="utf-8") as f:
        reference_tags = set(f.read().lower().split(", "))

    count_copied = 0
    total_imgs = 0
    for file in os.listdir(subfolder):
        if file.lower().endswith((".jpg",".png",".jpeg")):
            total_imgs += 1
            img_path = os.path.join(subfolder, file)
            img = Image.open(img_path).convert("RGBA")
            sorted_tags, _ = predictor.predict(img,general_thresh,character_thresh)
            tags_set = set(sorted_tags.lower().split(", "))
            if len(tags_set & reference_tags) >= min_similares:
                shutil.copy(img_path, similar_dir)
                count_copied +=1

    return f"‚úÖ Procesadas {total_imgs} im√°genes. {count_copied} copiadas a similar."

# Carpetas base y dropdown
base_path = "/content/drive/MyDrive/Capture"

# =========================
# Dropdown Carpeta y Subcarpeta
# =========================
def listar_carpeta_principal():
    if not os.path.exists(base_path):
        return gr.update(choices=[], value=None)
    folders = [f for f in sorted(os.listdir(base_path)) if os.path.isdir(os.path.join(base_path,f))]
    return gr.update(choices=folders, value=None)

def listar_subcarpetas(selected_carpeta):
    if not selected_carpeta:
        return gr.update(choices=[], value=None)
    path = os.path.join(base_path, selected_carpeta)
    if not os.path.exists(path):
        return gr.update(choices=[], value=None)
    subfolders = [f for f in sorted(os.listdir(path)) if os.path.isdir(os.path.join(path,f))]
    return gr.update(choices=subfolders, value=None)

def seleccionar_subcarpeta(selected_carpeta, selected_subcarpeta):
    if not selected_carpeta or not selected_subcarpeta:
        return "‚ö†Ô∏è No se seleccion√≥ subcarpeta", ""
    full_path = os.path.join(base_path, selected_carpeta, selected_subcarpeta)
    return f"‚úÖ Subcarpeta activa: {selected_subcarpeta}", full_path

# Autocompletado de Tags
csv_path = "/content/WaifuDiffusion/extension/tagcomplete/tags/danbooru.csv"
df_tags = pd.read_csv(csv_path, header=None, names=["name","type","postCount","aliases"])
def autocompletado(prefijo):
    pref = prefijo.lower().strip().replace(" ", "_")
    if pref == "":
        return []
    mask_name = df_tags['name'].str.lower().str.contains(pref)
    resultados = df_tags[mask_name].sort_values('postCount', ascending=False).head(5)['name'].tolist()
    return resultados

# --- Interfaz Gradio ---
with gr.Blocks(theme="lone17/kotaemon") as demo:
    gr.Markdown("# ü§ñ WaifuDiffusion")

    with gr.Row():
        with gr.Column(variant="panel"):
            image_in = gr.Image(type="filepath", label="Imagen", height=400)

            with gr.Accordion("Opciones", open=False):
                general_thresh = gr.Slider(0.1,0.9,0.5,step=0.1,label="Umbral General")
                character_thresh = gr.Slider(0.1,0.9,0.5,step=0.1,label="Umbral Personajes") # NUEVO
                min_similares = gr.Slider(1,10,2,step=1,label="M√≠nimo de tags similares a copiar")
                with gr.Accordion("Opciones avanzadas", open=False):
                    show_wiki = gr.Checkbox(label="Mostrar Wiki", value=False)

            # Botones: Procesar, Guardar (gris), Buscar personaje
            with gr.Row():
                process_btn = gr.Button("Procesar", variant="primary")
                save_btn = gr.Button("Guardar", variant="secondary")  # gris
                buscar_btn = gr.Button("Buscar personaje", variant="primary")

            # Accordion Carpetas
            with gr.Accordion("Carpetas", open=False):
                carpeta_dropdown = gr.Dropdown([], label="Seleccionar carpeta")
                subcarpeta_dropdown = gr.Dropdown([], label="Seleccionar subcarpeta")
                ruta_seleccionada = gr.State("")

        with gr.Column(variant="panel"):
            sorted_out = gr.Textbox(label="Editar tags", lines=3, interactive=True)
            with gr.Accordion("Wiki", open=False, visible=False) as wiki_box:
                wiki_link = gr.HTML("")
            with gr.Row():
                tag_prefix = gr.Textbox(label="Buscar tag",
                                        placeholder="Ingresar un tag",
                                        value="",
                                        scale=0.4,
                                        lines=9)
                with gr.Column():
                    sug_buttons = [gr.Button("", visible=False) for _ in range(5)]
            msg_out = gr.Textbox(label="Estado", interactive=False)

    # --- Conexiones ---
    demo.load(listar_carpeta_principal, outputs=[carpeta_dropdown])
    carpeta_dropdown.change(listar_subcarpetas, inputs=[carpeta_dropdown], outputs=[subcarpeta_dropdown])
    subcarpeta_dropdown.change(seleccionar_subcarpeta,
                               inputs=[carpeta_dropdown, subcarpeta_dropdown],
                               outputs=[msg_out, ruta_seleccionada])

    # Guardar tags de Editar tags
    def guardar_tags(sorted_tags):
        if not sorted_tags.strip():
            return "‚ö†Ô∏è No hay tags para guardar"
        with open(reference_path, "w", encoding="utf-8") as f:
            f.write(sorted_tags)
        return f"‚úÖ Tags guardados en {reference_path}"

    save_btn.click(guardar_tags, inputs=[sorted_out], outputs=[msg_out])

    buscar_btn.click(buscar_personaje,
                     inputs=[ruta_seleccionada, general_thresh, character_thresh, min_similares],
                     outputs=msg_out)

    # --- Procesar imagen cargada ---
    def procesar_imagen(image_path, general_thresh, character_thresh):
        if not image_path:
            return "", "‚ö†Ô∏è No se carg√≥ ninguna imagen"
        img = Image.open(image_path).convert("RGBA")
        tags_str, combined = predictor.predict(img, general_thresh, character_thresh)
        return tags_str, "‚úÖ Imagen procesada y etiquetas generadas"

    process_btn.click(procesar_imagen,
                      inputs=[image_in, general_thresh, character_thresh],
                      outputs=[sorted_out, msg_out])

    # Autocompletado
    def actualizar_botones(prefijo):
        resultados = autocompletado(prefijo)
        updates = []
        for i, b in enumerate(sug_buttons):
            if i < len(resultados):
                display_txt = resultados[i].replace("_"," ")
                updates.append(gr.update(value=display_txt, visible=True))
            else:
                updates.append(gr.update(visible=False))
        return updates
    tag_prefix.change(actualizar_botones, inputs=[tag_prefix], outputs=sug_buttons)

    def actualizar_link(prefijo, mostrar):
        if not mostrar:
            return "", gr.update(visible=False)
        prefijo = prefijo.strip().replace(" ", "_")
        if not prefijo:
            return "<p>‚ö†Ô∏è Ingresa un tag primero</p>", gr.update(visible=True)
        url = f"https://danbooru.donmai.us/wiki_pages/{prefijo}"
        return f'<a href="{url}" target="_blank">üåê {url}</a>', gr.update(visible=True)
    tag_prefix.change(fn=actualizar_link, inputs=[tag_prefix, show_wiki], outputs=[wiki_link, wiki_box])

    clicks_state = [gr.State(0) for _ in range(5)]
    for i, btn in enumerate(sug_buttons):
        def click_sugerencia(b, actual, mostrar, clicks, idx=i):
            btn_text = b.replace("_"," ").strip()
            clicks = (clicks or 0) + 1
            if clicks % 2 == 1:
                link, box_state = actualizar_link(btn_text, mostrar)
                return gr.update(), link, box_state, clicks
            else:
                actual = (actual or "").strip()
                if not actual:
                    nuevo = btn_text
                else:
                    nuevo = f"{actual}, {btn_text}"
                return nuevo, "", gr.update(visible=False), clicks
        btn.click(click_sugerencia,
                  inputs=[btn, sorted_out, show_wiki, clicks_state[i]],
                  outputs=[sorted_out, wiki_link, wiki_box, clicks_state[i]])

    def toggle_wiki(mostrar):
        return gr.update(visible=mostrar, open=mostrar)
    show_wiki.change(toggle_wiki, inputs=[show_wiki], outputs=[wiki_box])

demo.queue()

# Mostrar el contenedor final con link de la app
import re
import threading
import time
from IPython.display import clear_output, display
import ipywidgets as widgets

def launch_and_get_url(demo):
    import sys
    import io
    old_stdout = sys.stdout
    sys.stdout = mystdout = io.StringIO()
    t = threading.Thread(target=lambda: demo.launch(share=True, prevent_thread_lock=True))
    t.start()

    url = None
    while url is None:
        time.sleep(1)
        output = mystdout.getvalue()
        match = re.search(r"https://[a-z0-9]+\.gradio\.live", output)
        if match:
            url = match.group(0)
    sys.stdout = old_stdout
    return url

app_url = launch_and_get_url(demo)

clear_output()
contenedor_fi = widgets.VBox([
    widgets.HBox([widgets.Output(), widgets.Output()]),
    widgets.Output()])

with contenedor_fi.children[0].children[0]:
    display_inside_box(f"Listo", left="0px", right="0px", Wide="45px")
with contenedor_fi.children[0].children[1]:
    display_outside_box(f'<a href="{app_url}" target="_blank" style="text-decoration:none;">ü§ñ Buscar personaje</a>', left="0px")

# Mostrar el contenedor final
display(contenedor_fi)
````

````python
# Recorte de Im√°genes con YOLOv8

#@markdown ## **Recorte Autom√°tico**
#@markdown Este script utiliza **YOLOv8** para detectar objetos en im√°genes y generar recortes autom√°ticos.
#@markdown Incluye soporte para:
#@markdown Selecci√≥n de modelo de detecci√≥n rostros, anime, poses, etc.
#@markdown Ajuste de recorte con **relaci√≥n de aspecto fija** `2:2`, `2:3`, `3:2` o **libre** `off`

utilizar_gpu = False  #@param {type:"boolean"}

#@markdown Control de **posici√≥n del recorte** arriba, centro, abajo, izquierda, derecha
#@markdown Vista previa de detecciones y recortes antes de guardar
#@markdown Guardado de los recortes seleccionados en la carpeta de salida  <p>
#@markdown ![YOLOv8 API](https://img.shields.io/badge/YOLOv8_API-292929?style=for-the-badge&labelColor=%23005577)

import os, cv2, numpy as np, gradio as gr, shutil, random, string
from PIL import Image
from IPython.display import clear_output, display, HTML
import ipywidgets as widgets

# Instalar ultralytics si no existe
try: from ultralytics import YOLO
except ImportError:
    !pip install ultralytics --upgrade
    from ultralytics import YOLO

# Carpetas necesarias
model_yolo = "/content/yolov8/models"
images_yolo = "/content/yolov8/Images"
os.makedirs(model_yolo, exist_ok=True)
os.makedirs(images_yolo, exist_ok=True)
similar_folder = os.path.join(project_file, "Imagen")
output_folder = os.path.join(project_file, "dataset")

# urls de modelos YOLO preentrenados
model_urls = {
    "Anime": "https://huggingface.co/jags/yolov8_model_segmentation-set/resolve/main/Anime-yolov8-seg.pt",
    "Detect": "https://github.com/ultralytics/assets/releases/download/v8.3.0/yolo11x-seg.pt",
    "Face": "https://github.com/Fuyucch1/yolov8_animeface/releases/download/v1/yolov8x6_animeface.pt",
    "Pose": "https://github.com/ultralytics/assets/releases/download/v8.3.0/yolo11x-pose.pt"}

# Cache de modelos
model_cache = {}

# Funci√≥n para aplicar relaci√≥n de aspecto
def enforce_aspect_ratio(image, aspect="off", position="centro"):
    if aspect == "off":
        return image
    h, w, _ = image.shape
    target_ratio = {"2:3":2/3,"3:2":3/2,"2:2":1}.get(aspect, w/h)
    current_ratio = w/h
    if current_ratio > target_ratio:
        new_w = int(h*target_ratio)
        pos_map_x = {"arriba-izq":0,"centro-izq":0,"abajo-izq":0,
                     "arriba":(w-new_w)//2,"centro":(w-new_w)//2,"abajo":(w-new_w)//2,
                     "arriba-der":w-new_w,"centro-der":w-new_w,"abajo-der":w-new_w}
        start_x = pos_map_x.get(position,(w-new_w)//2)
        return image[:,start_x:start_x+new_w]
    else:
        new_h = int(w/target_ratio)
        pos_map_y = {"arriba-izq":0,"arriba":0,"arriba-der":0,
                     "centro-izq":(h-new_h)//2,"centro":(h-new_h)//2,"centro-der":(h-new_h)//2,
                     "abajo-izq":h-new_h,"abajo":h-new_h,"abajo-der":h-new_h}
        start_y = pos_map_y.get(position,(h-new_h)//2)
        return image[start_y:start_y+new_h,:]

# Cargar modelo YOLO con cache y GPU/CPU
def load_model(model_name):
    if model_name in model_cache:
        return model_cache[model_name]
    model_path = os.path.join(model_yolo, os.path.basename(model_urls[model_name]))
    if not os.path.exists(model_path):
        os.system(f"wget {model_urls[model_name]} -O {model_path}")
    device = 0 if utilizar_gpu else "cpu"
    model = YOLO(model_path).to(device)
    model_cache[model_name] = model
    return model

# Detectar y recortar objetos
def detect_and_crop(image_path, model_name="Detect", num_objects=5, conf=0.2, aspect="off", position="centro"):
    if not os.path.exists(image_path):
        # CORRECCI√ìN: devolver siempre 5 valores (como en retorno normal)
        return [], [], [], None, ""
    model = load_model(model_name)
    img = Image.open(image_path).convert("RGB")
    img_array = np.array(img)
    base_name = os.path.splitext(os.path.basename(image_path))[0]
    img_folder = os.path.join(images_yolo, base_name)
    os.makedirs(img_folder, exist_ok=True)
    results = model.predict(img_array, conf=conf)
    tracked_img = results[0].plot()
    tracked_img_pil = Image.fromarray(cv2.cvtColor(tracked_img, cv2.COLOR_BGR2RGB))
    cropped_paths, labels, previews = [], [], []
    for i, box in enumerate(results[0].boxes.xyxy[:num_objects]):
        x1, y1, x2, y2 = map(int, box.tolist())
        x1, y1, x2, y2 = max(0, x1), max(0, y1), min(img_array.shape[1], x2), min(img_array.shape[0], y2)
        crop = img_array[y1:y2, x1:x2]
        crop = enforce_aspect_ratio(crop, aspect, position)
        crop_path = os.path.join(img_folder, f"{base_name}_crop_{i+1}.jpg")
        cv2.imwrite(crop_path, cv2.cvtColor(crop, cv2.COLOR_RGB2BGR))
        cropped_paths.append(crop_path)
        labels.append(f"Imagen {i+1}")
        previews.append(Image.fromarray(crop))
    return cropped_paths, labels, previews, tracked_img_pil, base_name

# Guardar recortes seleccionados con numeraci√≥n basada en save_counter
def save_selected_increment(selected_labels, all_crops, all_labels, counter):
    if not selected_labels:
        return "‚ö†Ô∏è No seleccionaste recortes"
    saved = []
    current_num = counter
    for sel in selected_labels:
        idx = all_labels.index(sel)
        src = all_crops[idx]
        rand_letters = ''.join(random.choices(string.ascii_lowercase, k=3))
        filename = f"{current_num:02d}{rand_letters}.png"
        shutil.copy(src, os.path.join(output_folder, filename))
        saved.append(filename)
    return f"‚úÖ Guardados {len(saved)} recortes: {', '.join(saved)}"

# Lista de im√°genes
def get_image_list():
    return sorted([os.path.join(similar_folder,f) for f in os.listdir(similar_folder) if f.lower().endswith((".jpg",".jpeg",".png"))], key=lambda x: x.lower())

# Siguiente imagen simple y aumentar contador
def next_image_with_counter(index, counter):
    files = get_image_list()
    if not files:
        return None, 0, counter
    next_idx = (index + 1) % len(files)
    counter += 1
    return files[next_idx], next_idx, counter

# --- Interfaz Gradio ---
with gr.Blocks(theme="lone17/kotaemon") as demo:
    gr.Markdown("# **üé± Ultralytics YOLOv8**")

    with gr.Row():
        with gr.Column(variant="panel"):
            inp_image = gr.Image(type="filepath", label="Imagen", height=400, visible=False)
            tracked_image = gr.Image(label="rastreada", interactive=False, height=400, visible=True)

            with gr.Accordion("Opciones", open=False):
                num_objects = gr.Slider(2,8,5,step=1,label="M√°x. objetos")
                conf = gr.Slider(0.1,0.5,0.2,step=0.1,label="Umbral confianza")

                with gr.Accordion("Opciones de recorte", open=False):
                    model_name = gr.Dropdown(list(model_urls.keys()), value="Pose", label="Tipo de modelo")
                    aspect = gr.Dropdown(choices=["Formato retrato","Formato cuadrado","Formato original"], value="Formato retrato", label="Tipo de corte")
                    aspect_state = gr.State("Formato retrato")

                with gr.Accordion("Opciones avanzadas", open=False):
                    toggle_rastreada = gr.Checkbox(label="ver imagen original", value=False)

            with gr.Row():
                recargar_btn = gr.Button("Procesar", variant="primary")
                next_btn = gr.Button("Siguiente", variant="secondary")
                save_btn = gr.Button("Guardar", variant="primary")

            with gr.Accordion("Posici√≥n de corte", open=False):
                position_state = gr.State("centro")
                buttons=[("‚Üñ","arriba-izq"),("‚Üë","arriba"),("‚Üó","arriba-der"),
                         ("‚Üê","centro-izq"),("‚Ä¢","centro"),("‚Üí","centro-der"),
                         ("‚Üô","abajo-izq"),("‚Üì","centro"),("‚Üò","abajo-der")]

                with gr.Row():
                    rotate_btn = gr.Button("Rotar corte", elem_classes="tiny-arrow")

                def toggle_aspect(current_aspect):
                    if current_aspect == "Formato retrato":
                        return "Formato 3:2"
                    elif current_aspect == "Formato 3:2":
                        return "Formato retrato"
                    return current_aspect

                rotate_btn.click(fn=toggle_aspect, inputs=[aspect_state], outputs=[aspect_state])

                for i in range(0,9,3):
                    with gr.Row():
                        for label,val in buttons[i:i+3]:
                            btn = gr.Button(label, elem_classes="tiny-arrow")
                            btn.click(fn=lambda v=val: v, inputs=[], outputs=[position_state])

        with gr.Column(variant="panel"):
            with gr.Row():
                cropped_gallery = gr.CheckboxGroup(choices=[], label="recortes", scale=1)
                cropped_preview = gr.Gallery(label="Vista previa recortes", columns=4, height="auto", scale=2)

    msg_out = gr.Textbox(label="Estado", interactive=False)
    crops_state = gr.State()
    labels_state = gr.State()
    index_state = gr.State(0)
    save_counter = gr.State(1)

    def run_detection(image_path, model_name, num_objects, conf, aspect_value, position):
        aspect_map = {"Formato retrato":"2:3","Formato 3:2":"3:2","Formato original":"off","Formato cuadrado":"2:2"}
        real_aspect = aspect_map.get(aspect_value, "off")
        crops, labels, previews, tracked_img, base_name = detect_and_crop(image_path, model_name, num_objects, conf, real_aspect, position)
        estado = f"Imagen procesada: {base_name}"
        return gr.update(choices=labels, value=labels), previews, labels, crops, tracked_img, estado

    inp_image.change(fn=run_detection,
                     inputs=[inp_image, model_name, num_objects, conf, aspect_state, position_state],
                     outputs=[cropped_gallery,cropped_preview,labels_state,crops_state,tracked_image,msg_out])

    recargar_btn.click(fn=run_detection,
                       inputs=[inp_image, model_name, num_objects, conf, aspect_state, position_state],
                       outputs=[cropped_gallery,cropped_preview,labels_state,crops_state,tracked_image,msg_out])

    save_btn.click(save_selected_increment,
                   inputs=[cropped_gallery,crops_state,labels_state,save_counter],
                   outputs=[msg_out])

    aspect.change(lambda v: v, inputs=[aspect], outputs=[aspect_state])

    def toggle_rotate_btn(value):
        if value in ["Formato original","Formato cuadrado"]:
            return gr.update(visible=False)
        return gr.update(visible=True)

    aspect.change(toggle_rotate_btn, inputs=[aspect], outputs=[rotate_btn])

    # Alternar imagen rastreada / original
    def toggle_images_original(show_original):
        if show_original:
            return gr.update(visible=True), gr.update(visible=False)
        else:
            return gr.update(visible=False), gr.update(visible=True)

    toggle_rastreada.change(toggle_images_original, inputs=[toggle_rastreada], outputs=[inp_image, tracked_image])

    # Bot√≥n Siguiente incrementa contador
    next_btn.click(next_image_with_counter,
                   inputs=[index_state, save_counter],
                   outputs=[inp_image,index_state, save_counter])

    demo.load(lambda: (get_image_list()[0],0,1) if get_image_list() else (None,0,1),
              inputs=None, outputs=[inp_image,index_state, save_counter])

# CSS botones peque√±os
display(HTML("<style>.tiny-arrow .gr-button{font-size:12px!important;padding:2px 4px!important;width:32px!important;}</style>")); clear_output()

# Mostrar solo badge con link de la app
import re, threading, time, sys, io

def launch_and_get_url(demo):
    old_stdout = sys.stdout
    mystdout = io.StringIO()
    sys.stdout = mystdout
    threading.Thread(target=lambda: demo.launch(share=True, prevent_thread_lock=True)).start()
    url = None
    while url is None:
        time.sleep(1)
        match = re.search(r"https://[a-z0-9]+\.gradio\.live", mystdout.getvalue())
        if match: url = match.group(0)
    sys.stdout = old_stdout
    return url

# Obtener link de Gradio autom√°ticamente
app_url = launch_and_get_url(demo)

# Mostrar badge clicable
clear_output()
contenedor_fi = widgets.VBox([widgets.HBox([widgets.Output(), widgets.Output()]), widgets.Output()])
with contenedor_fi.children[0].children[0]: display_inside_box("Listo", left="0px", right="0px", Wide="45px")
with contenedor_fi.children[0].children[1]: display_outside_box(f'<a href="{app_url}" target="_blank" style="text-decoration:none;">üé± Ultralytics</a>', left="0px")
display(contenedor_fi)

#Fin
````

````python
# Etiquetado Autom√°tico con EVA02 Large Tagger

#@markdown ### **Etiquetador de Im√°genes**
#@markdown Este script utiliza el modelo **WaifuDiffusion** para analizar im√°genes y generar etiquetas autom√°ticamente.
#@markdown Incluye soporte para:
#@markdown Detecci√≥n de **personajes**, Etiquetas **generales** ropa, expresiones, objetos y m√°s. Botones de sugerencia con **doble acci√≥n** <p>

utilizar_gpu = True  #@param {type:"boolean"}
#@markdown Interfaz gr√°fica con **Gradio** para cargar im√°genes, visualizar etiquetas y editarlas f√°cilmente.
#@markdown Los resultados se muestran en formato legible (sin guiones bajos).

!pip install --upgrade gradio pandas -q
if utilizar_gpu:
    !pip install --upgrade onnxruntime-gpu -q
else:
    !pip install --upgrade onnxruntime -q

import os
import numpy as np
import pandas as pd
from PIL import Image
import gradio as gr
import onnxruntime as rt

# Clonar repositorio de tags en la carpeta indicada
if not os.path.exists("/content/WaifuDiffusion/extension/tagcomplete"):
    !git clone https://github.com/DominikDoom/a1111-sd-webui-tagcomplete.git /content/WaifuDiffusion/extension/tagcomplete

# Crear carpeta model y similar
model_dir = "/content/WaifuDiffusion/model"
os.makedirs(model_dir, exist_ok=True)

# Variables para las rutas visualizadas
model_path = "/content/WaifuDiffusion/model/model.onnx"
labels_path = "/content/WaifuDiffusion/model/selected_tags.csv"

# Descargar modelo y etiquetas si no existen
if not os.path.exists(model_path):
    os.system(f"wget -q https://huggingface.co/SmilingWolf/wd-eva02-large-tagger-v3/resolve/main/model.onnx -O {model_path}")
if not os.path.exists(labels_path):
    os.system(f"wget -q https://huggingface.co/SmilingWolf/wd-eva02-large-tagger-v3/resolve/main/selected_tags.csv -O {labels_path}")

kaomojis = ["0_0","(o)_(o)","+_+","+-_","._.","<o>_<o>","<|>_<|>","=_=",
            ">_<","3_3","6_9",">_o","@_@","^_^","o_o","u_u","x_x","|_|","||_||"]

def load_labels():
    df = pd.read_csv(labels_path)
    names = df["name"].map(lambda x: x.replace("_"," ") if x not in kaomojis else x).tolist()
    rating_idx = list(np.where(df["category"]==9)[0])
    general_idx = list(np.where(df["category"]==0)[0])
    character_idx = list(np.where(df["category"]==4)[0])
    return names, rating_idx, general_idx, character_idx

tag_names, rating_idx, general_idx, character_idx = load_labels()

# Cargar modelo EVA02 como predictor con opci√≥n GPU/CPU
class Tagger:
    def __init__(self, model_path, utilizar_gpu=True):
        providers = (["CUDAExecutionProvider", "CPUExecutionProvider"]
                     if utilizar_gpu else ["CPUExecutionProvider"])
        self.model = rt.InferenceSession(model_path, providers=providers)
        _, h, w, _ = self.model.get_inputs()[0].shape
        self.size = h

    def prepare(self, image):
        canvas = Image.new("RGBA", image.size, (255, 255, 255))
        canvas.alpha_composite(image)
        image = canvas.convert("RGB")
        max_dim = max(image.size)
        pad_left, pad_top = (max_dim - image.size[0]) // 2, (max_dim - image.size[1]) // 2
        padded = Image.new("RGB", (max_dim, max_dim), (255, 255, 255))
        padded.paste(image, (pad_left, pad_top))
        if max_dim != self.size:
            padded = padded.resize((self.size, self.size), Image.BICUBIC)
        arr = np.asarray(padded, dtype=np.float32)[:, :, ::-1]
        return np.expand_dims(arr, 0)

    def predict(self, image, general_thresh=0.5, character_thresh=0.5):
        img = self.prepare(image)
        input_name = self.model.get_inputs()[0].name
        preds = self.model.run(None, {input_name: img})[0][0].astype(float)

        # M√°s umbral = m√°s tags
        general_thresh, character_thresh = 1 - general_thresh, 1 - character_thresh
        characters = {tag_names[i]: preds[i] for i in character_idx if preds[i] > character_thresh}
        general = {tag_names[i]: preds[i] for i in general_idx if preds[i] > general_thresh}
        combined = (
            sorted(characters.items(), key=lambda x: x[1], reverse=True) +
            sorted(general.items(), key=lambda x: x[1], reverse=True))
        sorted_str = ", ".join(name for name, _ in combined)
        return sorted_str, combined

# Inicializar predictor con la opci√≥n elegida
predictor = Tagger(model_path, utilizar_gpu=utilizar_gpu)

# Carpeta para dataset de im√°genes
image_folder = os.path.join(project_file, "dataset")
os.makedirs(image_folder, exist_ok=True)
image_list = sorted([os.path.join(image_folder,f) for f in os.listdir(image_folder)
                     if f.lower().endswith((".jpg",".png",".jpeg"))])

# Funciones auxiliares
def load_image(idx):
    if not image_list:
        return None, 0, ""
    idx = idx % len(image_list)
    path = image_list[idx]
    txt_name = os.path.splitext(os.path.basename(path))[0]+".txt"
    txt_path = os.path.join(image_folder, txt_name)
    tags_str = ""
    if os.path.exists(txt_path):
        with open(txt_path, "r", encoding="utf-8") as f:
            tags_str = f.read()
    return path, idx, tags_str

def next_image(idx):
    return load_image(idx+1)

def process_image(image_path, general_thresh, character_thresh, idx, procesar_todo):
    if procesar_todo:
        if not image_list:
            return gr.update(), "", {}, idx, "‚ö†Ô∏è No hay im√°genes en la carpeta dataset"
        for img_path in image_list:
            img = Image.open(img_path).convert("RGBA")
            sorted_tags, _ = predictor.predict(img,general_thresh,character_thresh)
            txt_name = os.path.splitext(os.path.basename(img_path))[0]+".txt"
            txt_path = os.path.join(image_folder, txt_name)
            with open(txt_path, "w", encoding="utf-8") as f:
                f.write(sorted_tags)
        return image_path, "", {}, idx, f"‚úÖ Procesadas {len(image_list)} im√°genes y guardados sus .txt"
    else:
        if not image_path:
            return gr.update(), "", {}, idx, "‚ö†Ô∏è No hay imagen cargada"
        img = Image.open(image_path).convert("RGBA")
        sorted_tags, combined = predictor.predict(img,general_thresh,character_thresh)
        return image_path, sorted_tags, dict(combined), idx, "‚úÖ Imagen procesada"

def save_tags(image_path, edited_tags, agregar_personaje, personaje_principal):
    if agregar_personaje and personaje_principal.strip():
        # Reemplazar _ por espacio
        personaje = personaje_principal.strip().replace("_", " ")
        # Guardar al inicio en TODAS las im√°genes
        for img_path in image_list:
            txt_name = os.path.splitext(os.path.basename(img_path))[0]+".txt"
            txt_path = os.path.join(image_folder, txt_name)
            contenido = ""
            if os.path.exists(txt_path):
                with open(txt_path, "r", encoding="utf-8") as f:
                    contenido = f.read().strip()
            # Insertar personaje al inicio
            nuevo = f"{personaje}, {contenido}" if contenido else personaje
            with open(txt_path, "w", encoding="utf-8") as f:
                f.write(nuevo)
        return f"‚úÖ personaje agregado al inicio de todas las im√°genes"
    else:
        if not image_path:
            return "‚ö†Ô∏è No hay imagen cargada"
        txt_name = os.path.splitext(os.path.basename(image_path))[0]+".txt"
        txt_path = os.path.join(image_folder, txt_name)
        with open(txt_path, "w", encoding="utf-8") as f:
            f.write(edited_tags)
        return f"‚úÖ Tags guardados en {os.path.basename(txt_path)}"

# Autocompletado de Tags
csv_path = "/content/WaifuDiffusion/extension/tagcomplete/tags/danbooru.csv"
df_tags = pd.read_csv(csv_path, header=None, names=["name","type","postCount","aliases"])

def autocompletado(prefijo):
    pref = prefijo.lower().strip().replace(" ", "_")
    if pref == "":
        return []
    mask_name = df_tags['name'].str.lower().str.contains(pref)
    resultados = df_tags[mask_name].sort_values('postCount', ascending=False).head(5)['name'].tolist()
    return resultados

# Crear y organizar interfaz en Gradio
with gr.Blocks(theme="lone17/kotaemon") as demo:
    gr.Markdown("# üå∏ WaifuDiffusion")

    with gr.Row():
        with gr.Column(variant="panel"):
            image_in = gr.Image(type="filepath", label="Imagen", height=400)

            with gr.Accordion("Opciones", open=False):
                general_thresh = gr.Slider(0,1,0.5,step=0.1,label="Umbral General")
                character_thresh = gr.Slider(0,1,0.5,step=0.1,label="Umbral Personajes")

                with gr.Accordion("Opciones avanzadas", open=False):
                    show_wiki = gr.Checkbox(label="Mostrar Wiki", value=False)
                    procesar_todo = gr.Checkbox(label="Procesar todo", value=False)
                    agregar_personaje = gr.Checkbox(label="Agregar etiquetas a todo", value=False)

            with gr.Row():
                process_btn = gr.Button("Procesar", variant="primary")
                next_btn = gr.Button("Siguiente", variant="secondary")
                save_btn = gr.Button("Guardar", variant="primary")

        with gr.Column(variant="panel"):
            sorted_out = gr.Textbox(label="Editar tags", lines=3, interactive=True)

            # Crear espacio para Wiki
            with gr.Accordion("Wiki", open=False, visible=False) as wiki_box:
                wiki_link = gr.HTML("")

            # Crear contenedor de Prefijo y Sugerencias
            with gr.Row():
                tag_prefix = gr.Textbox(label="Agregar etiquetas",
                                        placeholder="Ingresar un tag",
                                        value="",
                                        scale=0.4,
                                        lines=8)
                with gr.Column():
                    sug_buttons = [gr.Button("", visible=False) for _ in range(5)]

            msg_out = gr.Textbox(label="Estado", interactive=False)

    index_state = gr.State(0)

    # Llenar los contenedores con el contenido correspondiente
    process_btn.click(process_image,
                      inputs=[image_in,general_thresh,character_thresh,index_state,procesar_todo],
                      outputs=[image_in, sorted_out, gr.State(), index_state, msg_out])

    next_btn.click(next_image,
                   inputs=[index_state],
                   outputs=[image_in,index_state,sorted_out])

    save_btn.click(save_tags,
                   inputs=[image_in,sorted_out,agregar_personaje,tag_prefix],
                   outputs=[msg_out])

    def actualizar_botones(prefijo):
        resultados = autocompletado(prefijo)
        updates = []
        for i, b in enumerate(sug_buttons):
            if i < len(resultados):
                display_txt = resultados[i].replace("_"," ")
                updates.append(gr.update(value=display_txt, visible=True))
            else:
                updates.append(gr.update(visible=False))
        return updates

    tag_prefix.change(actualizar_botones, inputs=tag_prefix, outputs=sug_buttons)

    def actualizar_link(prefijo, mostrar):
        if not mostrar:
            return "", gr.update(visible=False)
        prefijo = prefijo.strip().replace(" ", "_")
        if not prefijo:
            return "<p>‚ö†Ô∏è Ingresa un tag primero</p>", gr.update(visible=True)
        url = f"https://danbooru.donmai.us/wiki_pages/{prefijo}"
        return f'<a href="{url}" target="_blank">üåê {url}</a>', gr.update(visible=True)

    tag_prefix.change(fn=actualizar_link,
                      inputs=[tag_prefix, show_wiki],
                      outputs=[wiki_link, wiki_box])

    # Estado de clics para cada bot√≥n
    clicks_state = [gr.State(0) for _ in range(5)]

    for i, btn in enumerate(sug_buttons):
        def click_sugerencia(b, actual, mostrar, clicks, idx=i):
            btn_text = b.replace("_"," ").strip()
            clicks = (clicks or 0) + 1

            if clicks % 2 == 1:
                # Clic impar ‚Üí Wiki
                link, box_state = actualizar_link(btn_text, mostrar)
                return gr.update(), link, box_state, clicks
            else:
                # Clic par ‚Üí Editar tags
                actual = (actual or "").strip()
                if not actual:
                    nuevo = btn_text
                else:
                    nuevo = f"{actual}, {btn_text}"
                return nuevo, "", gr.update(visible=False), clicks

        btn.click(click_sugerencia,
                  inputs=[btn, sorted_out, show_wiki, clicks_state[i]],
                  outputs=[sorted_out, wiki_link, wiki_box, clicks_state[i]])

    def toggle_wiki(mostrar):
        return gr.update(visible=mostrar, open=mostrar)
    show_wiki.change(toggle_wiki, inputs=[show_wiki], outputs=[wiki_box])

    def load_first():
        return load_image(0)

    demo.load(load_first, outputs=[image_in,index_state,sorted_out])

demo.queue()

# Mostrar el contenedor final con link de la app
import re
import threading
import time
from IPython.display import clear_output, display
import ipywidgets as widgets

def launch_and_get_url(demo):
    import sys
    import io
    old_stdout = sys.stdout
    sys.stdout = mystdout = io.StringIO()
    t = threading.Thread(target=lambda: demo.launch(share=True, prevent_thread_lock=True))
    t.start()

    url = None
    while url is None:
        time.sleep(1)
        output = mystdout.getvalue()
        match = re.search(r"https://[a-z0-9]+\.gradio\.live", output)
        if match:
            url = match.group(0)
    sys.stdout = old_stdout
    return url

app_url = launch_and_get_url(demo)

clear_output()
contenedor_fi = widgets.VBox([
    widgets.HBox([widgets.Output(), widgets.Output()]),
    widgets.Output()])

with contenedor_fi.children[0].children[0]:
    display_inside_box(f"Listo", left="0px", right="0px", Wide="45px")
with contenedor_fi.children[0].children[1]:
    display_outside_box(f'<a href="{app_url}" target="_blank" style="text-decoration:none;">üå∏ WaifuDiffusion</a>', left="0px")

# Mostrar el contenedor final
display(contenedor_fi)
````

````python
# Mejorar im√°genes con Real-ESRGAN

import os
import shutil
from concurrent.futures import ThreadPoolExecutor, as_completed
from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets

# Mejorar im√°genes de Real-ESRGAN

#@markdown ### **Mejorar calidad de im√°genes**
#@markdown Este script permite mejorar la calidad de im√°genes utilizando el modelo Real-ESRGAN.
#@markdown Las im√°genes procesadas se almacenan en Google Drive, organizadas en carpetas espec√≠ficas, y se puede elegir si reemplazar las im√°genes originales tras el procesamiento.

remplazar_im√°genes = True  # @param {type:"boolean"}

#@markdown El usuario tambi√©n puede ajustar el nivel de **escalado** para aumentar la resoluci√≥n de las im√°genes.
escalado = 2  # @param {type:"slider", min:1, max:4, step:1}

# Verificar y clonar el repositorio si no existe
if not os.path.isdir('/content/Real-ESRGAN'):
    !git clone https://github.com/xinntao/Real-ESRGAN.git
os.chdir('/content/Real-ESRGAN')

# Instalar dependencias si no est√°n disponibles
try:
    import basicsr, facexlib, gfpgan, ffmpeg
except ImportError:
    !pip install basicsr facexlib gfpgan ffmpeg-python
    !pip install -r requirements.txt
    !python setup.py develop

# Modificar archivo para evitar errores
clear_output()
file_path = "/usr/local/lib/python3.11/dist-packages/basicsr/data/degradations.py"
with open(file_path, 'r') as file:
    file_data = file.read().replace('from torchvision.transforms.functional_tensor import rgb_to_grayscale', 'from torchvision.transforms.functional import rgb_to_grayscale')
with open(file_path, 'w') as file:
    file.write(file_data)

# Volver al directorio original
os.chdir('/content')

# Rutas y modelo (se mantienen iguales)
Esrgan_file = os.path.join(project_file, "dataset")
Esrgan_output = Esrgan_file
base_path = "/content/Real-ESRGAN"
model = "realesr-animevideov3"
code = f"{base_path}/inference_realesrgan.py"

# Filtrar im√°genes a procesar
image_files = [f for f in os.listdir(Esrgan_file) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Mostrar mensaje de inicio
clear_output()
display_outside_box("Procesando im√°genes", Wide="300px")

# Barra de progreso personalizada
progress_bar = display_ProgressBarAnimated(
    total=len(image_files),
    text="",
    width=320,
    height=30,
    color1="#3498db",  # Azul
    color2="#9b59b6")  # Morado

# Funci√≥n para mejorar una imagen usando GPU
def mejorar_imagen(filename):
    input_path = os.path.join(Esrgan_file, filename)
    command = f"python '{code}' -i '{input_path}' -n '{model}' -o '{Esrgan_output}' -s '{escalado}' --tile 200 --gpu-id 0 --fp32"
    os.system(command)
    progress_bar.update(1)  # Actualizar la barra de progreso

# Procesar im√°genes en paralelo con m√°s hilos para mayor velocidad
with ThreadPoolExecutor(max_workers=os.cpu_count() * 3) as executor:
    futures = {executor.submit(mejorar_imagen, filename): filename for filename in image_files}
    for future in as_completed(futures):
        filename = futures[future]
        try:
            future.result()
        except Exception as e:
            print(f"Error al procesar {filename}: {e}")

# Eliminar im√°genes originales si es necesario
if remplazar_im√°genes:
    for filename in image_files:
        original_path = os.path.join(Esrgan_file, filename)
        if os.path.exists(original_path):
            os.remove(original_path)

# Finalizar barra de progreso
progress_bar.complete("¬°Proceso completado!")

# Mensaje de finalizaci√≥n
clear_output()
total_images = len([f for f in os.listdir(Esrgan_output) if os.path.isfile(os.path.join(Esrgan_output, f))])
display_outside_box(f"Im√°genes procesadas: {total_images}", Wide="300px")

#Fin
````

````python

#@markdown ### **Ajustar tama√±o para Lora**
#@markdown Este script ajusta las im√°genes para que el lado m√°s corto siempre sea de **1000 p√≠xeles** (o **2000 p√≠xeles** si se selecciona HD), manteniendo la proporci√≥n original de la imagen.
#@markdown Si no se selecciona reemplazar, el nombre de la imagen se modifica con el sufijo `_re`.

remplazar_im√°genes = True  # @param {type:"boolean"}
resolucion_HD = False  # @param {type:"boolean"}

#@markdown El usuario puede elegir entre una resoluci√≥n est√°ndar (1000 p√≠xeles) o **HD** (2000 p√≠xeles) para el lado m√°s corto.
#@markdown Las im√°genes procesadas se almacenan en Google Drive, organizadas en carpetas espec√≠ficas. <br>

import os
import cv2
from IPython.display import display, clear_output
import ipywidgets as widgets

# Carpeta de entrada
Esrgan_file = os.path.join(project_file, "dataset")

# Definir lado corto objetivo
lado_corto_objetivo = 2000 if resolucion_HD else 1000  # 2000 para HD, 1000 para est√°ndar

# Obtener lista de im√°genes en la carpeta
image_files = [
    filename for filename in os.listdir(Esrgan_file)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')) and not filename.startswith('.')
]

# Mostrar mensaje de inicio
display_outside_box("Procesando im√°genes...", Wide="300px")

# Barra de progreso
progress = display_ProgressBarAnimated(
    total=len(image_files),
    text="",
    width=320,
    height=30,
    color1="#3498db",  # azul
    color2="#9b59b6"   # morado
)

# Funci√≥n para ajustar tama√±o
def ajustar_tama√±o(input_path, output_path):
    img = cv2.imread(input_path)
    if img is None:
        print(f"Error: No se pudo cargar la imagen {input_path}")
        return

    altura, ancho = img.shape[:2]

    # Escalado basado en el lado corto
    lado_corto = min(altura, ancho)
    factor_escalado = lado_corto_objetivo / lado_corto

    nuevo_ancho = int(ancho * factor_escalado)
    nueva_altura = int(altura * factor_escalado)

    img_escalada = cv2.resize(img, (nuevo_ancho, nueva_altura), interpolation=cv2.INTER_LANCZOS4)
    cv2.imwrite(output_path, img_escalada)

# Procesar im√°genes
for filename in image_files:
    input_path = os.path.join(Esrgan_file, filename)
    filename_sin_out = filename.replace('_out', '')

    if remplazar_im√°genes:
        output_path = input_path
    else:
        nombre, extension = os.path.splitext(filename_sin_out)
        output_path = os.path.join(Esrgan_file, f"{nombre}_re{extension}")

    ajustar_tama√±o(input_path, output_path)
    progress.update(step=1)

# Mensaje final
clear_output()
total_images = len(image_files)
if remplazar_im√°genes:
    display_outside_box(f"{total_images} im√°genes ajustadas y reemplazadas.", Wide="300px")
else:
    display_outside_box(f"{total_images} im√°genes ajustadas y guardadas.", Wide="300px")

# Fin
````

````python
import os, re, toml, pathlib, warnings
from time import time
from IPython.display import Markdown, display

warnings.filterwarnings("ignore")

# Variables de ejecuciones pasadas
old_model_url = model_url if "model_url" in globals() else None
dependencies_installed = globals().get("dependencies_installed", False)
model_file = globals().get("model_file", None)

# Variables heredadas o configurables
custom_dataset = globals().get("custom_dataset", None)
override_dataset_config_file = globals().get("override_dataset_config_file", None)
continue_from_lora = globals().get("continue_from_lora", "")
override_config_file = globals().get("override_config_file", None)

COLAB = True
SOURCE = "https://github.com/qaneel/kohya-trainer"
COMMIT = None
BETTER_EPOCH_NAMES = True
LOAD_TRUNCATED_IMAGES = True
try:
    LOWRAM = int(next(line.split()[1] for line in open('/proc/meminfo') if "MemTotal" in line)) / (1024**2) < 15
except:
    LOWRAM = False

#@markdown ### **Modelos**
folder_structure = "Organizar por proyecto (MyDrive/Loras/project/dataset)"
#@markdown Decide el modelo que se descargar√° y se usar√° para entrenamiento
modelo_de_entrenamiento = "WAI Illustrious v15" #@param ["Pony Diffusion V6 XL", "Animagine XL V3", "WAI Illustrious v15", "Stable Diffusion XL 1.0 base"]
training_model = modelo_de_entrenamiento
#@markdown Usar un modelo diffusers consume menos recursos. Las 3 opciones de arriba funcionar√°n con o sin diffusers.
load_diffusers = True
wandb_key = None

# Asignar modelo seg√∫n training_model y load_diffusers
if "Pony" in training_model:
    model_file = "/content/ponyDiffusionV6XL.safetensors"
    model_url = "https://huggingface.co/hollowstrawberry/67AB2F" if load_diffusers else "https://civitai.com/api/download/models/290640"

elif "Animagine" in training_model:
    model_file = "/content/animagineXLV3.safetensors"
    model_url = "https://huggingface.co/cagliostrolab/animagine-xl-3.0" if load_diffusers else "https://civitai.com/api/download/models/293564"

elif "WAI" in training_model:
    model_file = "/content/waiIllustriousV15.safetensors"
    model_url = ("https://huggingface.co/John6666/wai-nsfw-illustrious-sdxl-v150-sdxl"
                if load_diffusers else
                "https://civitai.com/api/download/models/2167369")

else:
    model_file = "/content/sdxl_base_1.0.safetensors"
    model_url = ("https://huggingface.co/stabilityai/stable-diffusion-xl-base-1.0/"
                if load_diffusers else
                "https://huggingface.co/stabilityai/stable-diffusion-xl-base-1.0/resolve/main/sd_xl_base_1.0.safetensors")

# Asignar VAE
vae_file = "stabilityai/sdxl-vae" if load_diffusers else "/content/sdxl_vae.safetensors"
if not load_diffusers: vae_url = "https://huggingface.co/stabilityai/sdxl-vae/resolve/main/sdxl_vae.safetensors"

# Limpiar URL
model_url = model_url.strip()
project_name = ruta_n
#@markdown ### **Procesamiento**
resolution = 1024 #param {type:"slider", min:768, max:1536, step:128}
caption_extension = ".txt"  # extensi√≥n de los captions
#@markdown Mezclar etiquetas de anime en su lugar mejora el aprendizaje y el prompting. Una etiqueta de activaci√≥n va al inicio de cada archivo de texto y no se mezclar√°.<p>
shuffle_tags = True
shuffle_caption = shuffle_tags
etiquetas_de_activaci√≥n = "1" #@param [0,1,2,3]
activation_tags = etiquetas_de_activaci√≥n
keep_tokens = int(activation_tags)

#@markdown ### **Pasos** <p>
#@markdown Tus im√°genes se repetir√°n durante el entrenamiento. Recomiendo que sean alrededor de 50 repeticiones.
repeticiones = 10 #@param {type:"number"}
num_repeats = repeticiones
#@markdown ### **√âpocas**
#@markdown Elige cu√°nto tiempo quieres entrenar. Un buen punto de partida es alrededor de 10 √©pocas o 500 pasos.<p>
#@markdown Una √©poca es un n√∫mero de pasos igual a: tu n√∫mero de im√°genes multiplicado por sus repeticiones, dividido por el tama√±o del batch.<p>
preferred_unit = "√âpocas"
cantidad_Loras = 5 #@param {type:"slider", min:1, max:10, step:1}
how_many = cantidad_Loras
max_train_epochs = how_many if preferred_unit == "√âpocas" else None
max_train_steps = how_many if preferred_unit == "Pasos" else None
#@markdown Guardar m√°s √©pocas te permitir√° comparar mejor el progreso de tu LoRA.
save_every_n_epochs = 1
keep_only_last_n_epochs = cantidad_Loras
if not save_every_n_epochs:
  save_every_n_epochs = max_train_epochs
if not keep_only_last_n_epochs:
  keep_only_last_n_epochs = max_train_epochs

#@markdown ### **Aprendizaje**
# La tasa de aprendizaje es lo m√°s importante para tus resultados.
# Si tu LoRA produce im√°genes negras, reduce UNet y Text Encoder a 1e-4 y 1e-5 respectivamente
# Si est√°s entrenando un estilo, puedes elegir poner el Text Encoder en 0.
unet_lr = 3e-4
text_encoder_lr = 6e-5
#@markdown El scheduler es el algoritmo que gu√≠a la tasa de aprendizaje. Si no est√°s seguro, elige `constant` y ignora el n√∫mero. Recomiendo `cosine_with_restarts` con 3 reinicios.
scheduler = "cosine_with_restarts" #@param ["constant", "cosine", "cosine_with_restarts", "constant_with_warmup", "linear", "polynomial"]
lr_scheduler = scheduler
reinicios = 3 #@param {type:"number"}
lr_scheduler_number = reinicios
# Pasos para "calentar" la tasa de aprendizaje durante el entrenamiento para eficiencia. Recomiendo dejarlo en 5%.
lr_warmup_ratio = 0.1
lr_warmup_steps = 0
#@markdown Ajusta la p√©rdida con el tiempo, haciendo el aprendizaje m√°s eficiente. El paper recomienda 5.0, recomiendo 8.0 para anime. Valores mayores lo hacen menos estricto. Pon 0 para desactivar.
aprendizaje = 5 #@param {type:"slider", min:0.0, max:10.0, step:0.5}
min_snr_gamma = aprendizaje
#@markdown Multinoise puede ayudar con el balance de color (oscuros m√°s oscuros, claros m√°s claros).
multinoise = True #@param {type:"boolean"}

#@markdown ###  **Estructura**
#@markdown LoRA es el tipo cl√°sico y bueno para una variedad de prop√≥sitos. LoCon es bueno con estilos art√≠sticos ya que tiene m√°s capas para aprender m√°s aspectos del dataset.
tipo_de_lora = "LoRA" #@param ["LoRA", "LoCon"]
lora_type = tipo_de_lora

#@markdown A continuaci√≥n algunos valores recomendados para XL seg√∫n la configuraci√≥n:

#@markdown | tipo |  network_dim  |  network_alpha  |
#@markdown | :------: | :------: | :------: |
#@markdown | LoRA regular |  8  |  4  |
#@markdown | LoCon estilo |  16  |  8  |

#@markdown M√°s dim significa LoRA m√°s grande, puede almacenar m√°s informaci√≥n pero m√°s no siempre es mejor.
# Slider combinado para network_dim y conv_dim
network_dim = 16 #@param {type:"slider", min:1, max:32, step:1}
conv_dim = network_dim
# Slider combinado para network_alpha y conv_alpha
network_alpha = 32 #@param {type:"slider", min:1, max:32, step:1}
conv_alpha = network_alpha

network_module = "networks.lora"
network_args = None
if lora_type.lower() == "locon":
  network_args = [f"conv_dim={conv_dim}", f"conv_alpha={conv_alpha}"]

#@markdown ### **Entrenamiento**
#@markdown Ajusta estos par√°metros seg√∫n la configuraci√≥n de tu Colab.
#@markdown Si est√°s usando la versi√≥n gratuita, deber√≠as seleccionar un modelo diffusers al inicio de esta celda.
#@markdown
#@markdown Un batch size m√°s alto suele ser m√°s r√°pido pero usa m√°s memoria.
tama√±o_de_entrenamiento = 5 #@param {type:"slider", min:1, max:16, step:1}
train_batch_size = tama√±o_de_entrenamiento
#@markdown No he encontrado diferencia sustancial entre sdpa y xformers.
atenci√≥n_cruzada = "sdpa" #@param ["sdpa", "xformers"]
cross_attention = atenci√≥n_cruzada
mixed_precision = "fp16"
# Cachear los latentes en Drive a√±adir√° un archivo de 250KB junto a cada imagen pero usar√° considerablemente menos memoria.
cache_latents = True
cache_latents_to_drive = True
#@markdown La siguiente opci√≥n desactivar√° shuffle_tags y deshabilitar√° el entrenamiento del text encoder.
almacenar_salidas_codificador_texto = False  # @param {type:"boolean"}
cache_text_encoder_outputs = almacenar_salidas_codificador_texto

#@markdown ### **Avanzado**
#@markdown El optimizador es el algoritmo usado para entrenar. AdamW8Bit es el predeterminado y funciona muy bien, mientras que Prodigy ajusta autom√°ticamente la tasa de aprendizaje y puede tener varias ventajas como entrenar m√°s r√°pido debido a necesitar menos pasos, adem√°s de funcionar mejor para datasets peque√±os.
optimizador = "AdamW8bit" #@param ["AdamW8bit", "Prodigy", "DAdaptation", "DadaptAdam", "DadaptLion", "AdamW", "Lion", "SGDNesterov", "SGDNesterov8bit", "AdaFactor"]
optimizer = optimizador
# Argumentos recomendados para AdamW8bit: `weight_decay=0.1 betas=[0.9,0.99]`<p>
# Argumentos recomendados para Prodigy: `decouple=True weight_decay=0.01 betas=[0.9,0.999] d_coef=2 use_bias_correction=True safeguard_warmup=True`<p>
# Argumentos recomendados para CAME: `weight_decay=0.04`
optimizer_args = "weight_decay=0.1 betas=[0.9,0.99]"
optimizer_args = [a.strip() for a in optimizer_args.split(' ') if a]
# Si Dadapt o Prodigy son seleccionados y la casilla recomendada est√° marcada, los siguientes valores recomendados sobrescribir√°n cualquier configuraci√≥n previa:<p>
# `unet_lr=0.75`, `text_encoder_lr=0.75`, `network_alpha=network_dim`<p>
# Si CAME + REX es seleccionado y la casilla recomendada est√° marcada, los siguientes valores recomendados sobrescribir√°n cualquier configuraci√≥n previa:<p>
# `unet_lr=1e-4`, `text_encoder_lr=1e-6, betas=default_came_betas`
recommended_values = False

if any(opt in optimizer.lower() for opt in ["dadapt", "prodigy"]):
    if recommended_values:
        unet_lr = text_encoder_lr = 0.75
        network_alpha = network_dim
elif "CAME" in optimizer:
    optimizer, lr_scheduler = "CAME", "REX"
    if recommended_values:
        unet_lr, text_encoder_lr = 1e-4, 1e-6
        optimizer_args = [arg for arg in optimizer_args if "betas" not in arg]

lr_scheduler_num_cycles = lr_scheduler_number if lr_scheduler == "cosine_with_restarts" else 0
lr_scheduler_power = lr_scheduler_number if lr_scheduler == "polynomial" else 0

# üë©‚Äçüíª Aqu√≠ va el c√≥digo principal

root_dir = "/content" if COLAB else pathlib.Path.home() / "Loras"
deps_dir = os.path.join(root_dir, "deps")
repo_dir = os.path.join(root_dir, "kohya-trainer")

main_dir      = os.path.join(root_dir, "drive/MyDrive") if COLAB else root_dir
log_folder    = os.path.join(main_dir, "models", "logs")
config_folder = os.path.join(main_dir, "models", "logs", "config")
images_folder = os.path.join(main_dir, "Loras", ruta_c, project_name, "dataset")
output_folder = os.path.join(main_dir, "models", ruta_c, project_name)

config_file = os.path.join(config_folder, "training_config.toml")
dataset_config_file = os.path.join(config_folder, "dataset_config.toml")
accelerate_config_file = os.path.join(repo_dir, "accelerate_config/config.yaml")

def install_dependencies():
  os.chdir(root_dir)
  !git clone {SOURCE} {repo_dir}
  os.chdir(repo_dir)
  if COMMIT:
    !git reset --hard {COMMIT}
  !wget https://raw.githubusercontent.com/hollowstrawberry/kohya-colab/main/train_network_xl_wrapper.py -q -O train_network_xl_wrapper.py
  !wget https://raw.githubusercontent.com/hollowstrawberry/kohya-colab/main/dracula.py -q -O dracula.py

  !apt -y update -qq
  !apt -y install aria2 -qq
  !export RUSTFLAGS="-A invalid_reference_casting"
  !pip install accelerate==0.19.0 transformers==4.30.2 diffusers==0.18.2 \
    bitsandbytes==0.45.1 flax==0.7.5 opencv-python jax==0.4.23 jaxlib==0.4.23 \
    pytorch-lightning==1.9.0 voluptuous==0.13.1 toml==0.10.2 ftfy==6.1.1 einops==0.6.0 \
    safetensors pygments huggingface-hub==0.22.0 wandb invisible-watermark==0.2.0 open-clip-torch==2.20.0 \
    dadaptation==3.1 prodigyopt==1.0 lion-pytorch==0.1.2 \
    torch==2.5.1+cu121 --extra-index https://download.pytorch.org/whl/cu121
  !pip install -e .
  if cross_attention == "xformers":
    !pip install -q xformers
  if "CAME" in optimizer:
    !pip install came-pytorch
    !wget https://raw.githubusercontent.com/hollowstrawberry/kohya-colab/main/train_util.py -q -O library/train_util.py

  # parchear kohya para peque√±os ajustes
  if LOWRAM:
    !sed -i "s@cpu@cuda@" library/model_util.py
  if LOAD_TRUNCATED_IMAGES:
    !sed -i 's/from PIL import Image/from PIL import Image, ImageFile\nImageFile.LOAD_TRUNCATED_IMAGES=True/g' library/train_util.py # corregir error de jpegs truncados
  if BETTER_EPOCH_NAMES:
    !sed -i 's/{:06d}/{:02d}/g' library/train_util.py # hacer nombres de √©poca m√°s cortos
    !sed -i 's/"." + args.save_model_as)/"-{:02d}.".format(num_train_epochs) + args.save_model_as)/g' train_network.py # el nombre de la √∫ltima √©poca coincidir√° con las dem√°s

  from accelerate.utils import write_basic_config
  if not os.path.exists(accelerate_config_file):
    write_basic_config(save_location=accelerate_config_file)

  os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
  os.environ["BITSANDBYTES_NOWELCOME"] = "1"
  os.environ["SAFETENSORS_FAST_GPU"] = "1"
  os.environ["PYTHONWARNINGS"] = "ignore"

def validate_dataset():
  global lr_warmup_steps, lr_warmup_ratio, caption_extension, keep_tokens, model_url
  supported_types = (".png", ".jpg", ".jpeg", ".webp", ".bmp")

  if model_url.startswith("/content/drive/") and not os.path.exists(model_url):
    print("üí• Error: El modelo de entrenamiento personalizado que especificaste no se encontr√≥ en tu Google Drive.")
    return

  print("\nüíø Verificando dataset...")
  if not project_name.strip() or any(c in project_name for c in " .()\"'\\/"):
    print("üí• Error: Por favor, elige un nombre de proyecto v√°lido.")
    return

  # Encontrar las carpetas y archivos
  if custom_dataset:
    try:
      datconf = toml.loads(custom_dataset)
      datasets = [d for d in datconf["datasets"][0]["subsets"]]
    except:
      print(f"üí• Error: Tu dataset personalizado es inv√°lido o contiene un error. Revisa la plantilla original.")
      return
    reg = [d.get("image_dir") for d in datasets if d.get("is_reg", False)]
    datasets_dict = {d["image_dir"]: d["num_repeats"] for d in datasets}
    folders = datasets_dict.keys()
    files = [f for folder in folders for f in os.listdir(folder)]
    images_repeats = {folder: (len([f for f in os.listdir(folder) if f.lower().endswith(supported_types)]), datasets_dict[folder]) for folder in folders}
  else:
    reg = []
    folders = [images_folder]
    files = os.listdir(images_folder)
    images_repeats = {images_folder: (len([f for f in files if f.lower().endswith(supported_types)]), num_repeats)}

# Validaci√≥n
  for folder in folders:
    if not os.path.exists(folder):
      print(f"üí• Error: La carpeta {folder.replace('/content/drive/', '')} no existe."); return

  for folder, (img, _) in images_repeats.items():
    if not img:
      print(f"üí• Error: Tu carpeta {folder.replace('/content/drive/', '')} est√° vac√≠a."); return

  test_files = []
  for f in files:
    if not f.lower().endswith((caption_extension, ".npz") + supported_types):
      print(f"üí• Error: Archivo inv√°lido en el dataset: \"{f}\". Abortando."); return
    if any(f.endswith(supported_types) and ff.endswith(supported_types)
           and os.path.splitext(f)[0] == os.path.splitext(ff)[0]
           for ff in test_files):
      print(f"üí• Error: Los archivos {f} y {ff} no pueden tener el mismo nombre. Abortando."); return
    test_files.append(f)

  if caption_extension and not any(txt.lower().endswith(caption_extension) for txt in files):
    caption_extension = ""
  if continue_from_lora and not (continue_from_lora.endswith(".safetensors") and os.path.exists(continue_from_lora)):
    print("üí• Error: Ruta inv√°lida para un Lora existente. Ejemplo: /content/drive/MyDrive/Loras/example.safetensors"); return

# Cosas bonitas
  pre_steps_per_epoch = sum(img*rep for (img, rep) in images_repeats.values())
  steps_per_epoch = pre_steps_per_epoch/train_batch_size
  total_steps = max_train_steps or int(max_train_epochs*steps_per_epoch)
  estimated_epochs = int(total_steps/steps_per_epoch)
  lr_warmup_steps = int(total_steps*lr_warmup_ratio)

  for folder, (img, rep) in images_repeats.items():
    print("üìÅ"+folder.replace("/content/drive/", "") + (" (Regularizaci√≥n)" if folder in reg else ""))
    print(f"üìà Se encontraron {img} im√°genes con {rep} repeticiones, igualando {img*rep} pasos.")
  print(f"üìâ Divide {pre_steps_per_epoch} pasos entre batch size {train_batch_size} para obtener {steps_per_epoch} pasos por √©poca.")
  if max_train_epochs:
    print(f"üîÆ Habr√° {max_train_epochs} √©pocas, para alrededor de {total_steps} pasos totales de entrenamiento.")
  else:
    print(f"üîÆ Habr√° {total_steps} pasos, divididos en {estimated_epochs} √©pocas y algunos m√°s.")

  if total_steps > 10000:
    print("üí• Error: El total de pasos es demasiado alto. Probablemente cometiste un error. Abortando...")
    return

  return True

def create_config():
  global dataset_config_file, config_file, model_file

  if override_config_file:
    config_file = override_config_file
    print(f"\n‚≠ï Usando archivo de configuraci√≥n personalizado {config_file}")
  else:
    config_dict = {
      "network_arguments": {
        "unet_lr": unet_lr,
        "text_encoder_lr": text_encoder_lr if not cache_text_encoder_outputs else 0,
        "network_dim": network_dim,
        "network_alpha": network_alpha,
        "network_module": network_module,
        "network_args": network_args,
        "network_train_unet_only": text_encoder_lr == 0 or cache_text_encoder_outputs,
        "network_weights": continue_from_lora if continue_from_lora else None
      },
      "optimizer_arguments": {
        "learning_rate": unet_lr,
        "lr_scheduler": lr_scheduler,
        "lr_scheduler_num_cycles": lr_scheduler_num_cycles if lr_scheduler == "cosine_with_restarts" else None,
        "lr_scheduler_power": lr_scheduler_power if lr_scheduler == "polynomial" else None,
        "lr_warmup_steps": lr_warmup_steps if lr_scheduler != "constant" else None,
        "optimizer_type": optimizer,
        "optimizer_args": optimizer_args if optimizer_args else None,
      },
      "training_arguments": {
        "pretrained_model_name_or_path": model_file,
        "vae": vae_file,
        "max_train_steps": max_train_steps,
        "max_train_epochs": max_train_epochs,
        "train_batch_size": train_batch_size,
        "seed": 42,
        "max_token_length": 225,
        "xformers": cross_attention == "xformers",
        "sdpa": cross_attention == "sdpa",
        "min_snr_gamma": min_snr_gamma if min_snr_gamma > 0 else None,
        "lowram": LOWRAM,
        "no_half_vae": True,
        "gradient_checkpointing": True,
        "gradient_accumulation_steps": 1,
        "max_data_loader_n_workers": 8,
        "persistent_data_loader_workers": True,
        "mixed_precision": mixed_precision,
        "full_bf16": mixed_precision == "bf16",
        "cache_latents": cache_latents,
        "cache_latents_to_disk": cache_latents_to_drive,
        "cache_text_encoder_outputs": cache_text_encoder_outputs,
        "min_timestep": 0,
        "max_timestep": 1000,
        "prior_loss_weight": 1.0,
        "multires_noise_iterations": 6 if multinoise else None,
        "multires_noise_discount": 0.3 if multinoise else None,
      },
      "saving_arguments": {
        "save_precision": "fp16",
        "save_model_as": "safetensors",
        "save_every_n_epochs": save_every_n_epochs,
        "save_last_n_epochs": keep_only_last_n_epochs,
        "output_name": project_name,
        "output_dir": output_folder,
        "log_prefix": project_name,
        "logging_dir": log_folder,
        "wandb_api_key": None,
        "log_with": None,
      }
    }

    for key in config_dict:
      if isinstance(config_dict[key], dict):
        config_dict[key] = {k: v for k, v in config_dict[key].items() if v is not None}

    with open(config_file, "w") as f:
      f.write(toml.dumps(config_dict))
    print(f"\nüìÑ Configuraci√≥n guardada en {config_file}")

  if override_dataset_config_file:
    dataset_config_file = override_dataset_config_file
    print(f"‚≠ï Usando archivo de configuraci√≥n de dataset personalizado {dataset_config_file}")
  else:
    dataset_config_dict = {
      "general": {
        "resolution": resolution,
        "shuffle_caption": shuffle_caption and not cache_text_encoder_outputs,
        "keep_tokens": keep_tokens,
        "flip_aug": False,
        "caption_extension": caption_extension,
        "enable_bucket": True,
        "bucket_no_upscale": False,
        "bucket_reso_steps": 64,
        "min_bucket_reso": 256,
        "max_bucket_reso": 4096,
      },
      "datasets": toml.loads(custom_dataset)["datasets"] if custom_dataset else [
        {
          "subsets": [
            {
              "num_repeats": num_repeats,
              "image_dir": images_folder,
              "class_tokens": None if caption_extension else project_name
            }
          ]
        }
      ]
    }

    for key in dataset_config_dict:
      if isinstance(dataset_config_dict[key], dict):
        dataset_config_dict[key] = {k: v for k, v in dataset_config_dict[key].items() if v is not None}

    with open(dataset_config_file, "w") as f:
      f.write(toml.dumps(dataset_config_dict))
    print(f"üìÑ Configuraci√≥n del dataset guardada en {dataset_config_file}")

def download_model():
  global old_model_url, model_url, model_file, vae_url, vae_file
  real_model_url = model_url  # Hab√≠a una raz√≥n para tener una variable separada, pero la olvid√©.

  if real_model_url.startswith("/content/drive/"):
    # Modelo local, ya se verific√≥ que existe
    model_file = real_model_url
    print(f"üìÅ Usando archivo de modelo local: {model_file}")
    # Validaci√≥n
    if model_file.lower().endswith(".safetensors"):
      from safetensors.torch import load_file as load_safetensors
      try:
        test = load_safetensors(model_file)
        del test
      except:
        return False
    elif model_file.lower().endswith(".ckpt"):
      from torch import load as load_ckpt
      try:
        test = load_ckpt(model_file)
        del test
      except:
        return False
    return True

  else:

    # Modelo descargable
    if load_diffusers:
      if 'huggingface.co' in real_model_url:
          match = re.search(r'huggingface\.co/([^/]+)/([^/]+)', real_model_url)
          if match:
              username = match.group(1)
              model_name = match.group(2)
              model_file = f"{username}/{model_name}"
              from huggingface_hub import HfFileSystem
              fs = HfFileSystem()
              existing_folders = set(fs.ls(model_file, detail=False))
              necessary_folders = [ "scheduler", "text_encoder", "text_encoder_2", "tokenizer", "tokenizer_2", "unet", "vae" ]
              if all(f"{model_file}/{folder}" in existing_folders for folder in necessary_folders):
                print("üçÉ Modelo Diffusers identificado.")  # Ser√° manejado por kohya
                return True
      raise ValueError("üí• Error al cargar el modelo Diffusers. Si este modelo no es Diffusers, ¬øhas intentado desactivarlo en la parte superior del Colab?")

    # Definir nombre de archivo local
    if not model_file:
      if real_model_url.lower().endswith((".ckpt", ".safetensors")):
        model_file = f"/content{real_model_url[real_model_url.rfind('/'):]}"
      else:
        model_file = "/content/downloaded_model.safetensors"
        if os.path.exists(model_file):
          !rm "{model_file}"

    # HuggingFace
    if m := re.search(r"(?:https?://)?(?:www\.)?huggingface\.co/[^/]+/[^/]+/blob", real_model_url):
      real_model_url = real_model_url.replace("blob", "resolve")
    # Civitai
    elif m := re.search(r"(?:https?://)?(?:www\\.)?civitai\.com/models/([0-9]+)(/[A-Za-z0-9-_]+)?", real_model_url):
      if m.group(2):
        model_file = f"/content{m.group(2)}.safetensors"
      if m := re.search(r"modelVersionId=([0-9]+)", real_model_url):
        real_model_url = f"https://civitai.com/api/download/models/{m.group(1)}"
      else:
        raise ValueError("üí• optional_custom_training_model contiene un enlace de Civitai, pero no incluye un modelVersionId. Tambi√©n puedes hacer clic derecho en el bot√≥n de descarga y copiar el enlace directo.")

    # Descargar checkpoint
    !aria2c "{real_model_url}" --console-log-level=warn -c -s 16 -x 16 -k 10M -d / -o "{model_file}"

    # Descargar VAE
    if not os.path.exists(vae_file):
      !aria2c "{vae_url}" --console-log-level=warn -c -s 16 -x 16 -k 10M -d / -o "{vae_file}"

    # Validaci√≥n

    if model_file.lower().endswith(".safetensors"):
      from safetensors.torch import load_file as load_safetensors
      try:
        test = load_safetensors(model_file)
        del test
      except:
        new_model_file = os.path.splitext(model_file)[0]+".ckpt"
        !mv "{model_file}" "{new_model_file}"
        model_file = new_model_file
        print(f"Renombrado modelo a {os.path.splitext(model_file)[0]}.ckpt")

    if model_file.lower().endswith(".ckpt"):
      from torch import load as load_ckpt
      try:
        test = load_ckpt(model_file)
        del test
      except:
        return False

  return True

def main():
  global dependencies_installed

  if COLAB and not os.path.exists('/content/drive'):
    from google.colab import drive
    print("üìÇ Conectando a Google Drive...")
    drive.mount('/content/drive')

  for dir in (main_dir, deps_dir, repo_dir, log_folder, images_folder, output_folder, config_folder):
    os.makedirs(dir, exist_ok=True)

  if not validate_dataset():
    return

  if not dependencies_installed:
    print("\nüè≠ Instalando dependencias...\n")
    t0 = time()
    install_dependencies()
    t1 = time()
    dependencies_installed = True
    print(f"\n‚úÖ Instalaci√≥n finalizada en {int(t1-t0)} segundos.")
  else:
    print("\n‚úÖ Dependencias ya instaladas.")

  if old_model_url != model_url or not model_file or not os.path.exists(model_file):
    print("\nüîÑ Obteniendo modelo...")
    if not download_model():
      print("\nüí• Error: El modelo especificado es inv√°lido o est√° corrupto."
            "\nSi est√°s usando una URL, verifica que sea accesible sin iniciar sesi√≥n."
            "\nPuedes usar URLs de Civitai o HuggingFace, o una ruta en tu Google Drive que comience con /content/drive/MyDrive")
      return
    print()
  else:
    print("\nüîÑ Modelo ya descargado.\n")

  create_config()

  print("\n‚≠ê Iniciando entrenador...\n")
  os.chdir(repo_dir)

  !accelerate launch --quiet --config_file={accelerate_config_file} --num_cpu_threads_per_process=1 train_network_xl_wrapper.py --dataset_config={dataset_config_file} --config_file={config_file}

  if not get_ipython().__dict__['user_ns']['_exit_code']:
    display(Markdown("### ‚úÖ ¬°Listo! [Ve a descargar tu Lora desde Google Drive](https://drive.google.com/drive/my-drive)\n"
                     "### Habr√° varios archivos, deber√≠as probar la √∫ltima versi√≥n (el archivo con el n√∫mero m√°s grande junto a √©l)"))

main()
````
